```r
#Part3 : ILI State Activity Indicatot Map Using Twitter Data

library(ggplot2)
library(maps)


US_states <- map_data("state")
#Load the file "ILI State Activity Indicatot Map Using Twitter Data"
twitter_data<-read.csv(file.choose(), header=T)
state_data<-aggregate(ILI.Activity.Level~region,FUN=length,data=twitter_data)

merged_data<- merge(US_states,state_data, by="region")

ggplot(merged_data, aes(x=long,y=lat,group=group))+geom_polygon(aes(fill=ILI.Activity.Level))+
scale_fill_gradientn(colours=c("green","yellow","orange","orangered","tomato","red","firebrick"),
label=c("minimal","low","moderate","High"),breaks=c(200,650,1300,1900), na.value="grey50")+
geom_path()+coord_map()+ggtitle("2017-2018 Season Week 4 ending Jan 27,2018(Using Twitter Data)")



```
